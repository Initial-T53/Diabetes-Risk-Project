```python
# Import libraries
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from statsmodels.stats.outliers_influence import variance_inflation_factor
from pathlib import Path
```

```python
#Load Datasets

diabetes=pd.read_csv('../dataset/diabetes_cleaned.csv')
```

```python
diabetes

# Testing to see ensure the dataset loaded properly
```

```python
X=diabetes.drop(['diabetes_risk_score','diabetes_stage', 'diagnosed_diabetes'],axis=1) # Removing the target variable and risk score (derived of other variable) and diabetes stage
y= diabetes['diagnosed_diabetes']                     # Target Variable
```

```python
X_train, X_test, y_train , y_test= train_test_split(
    X, y, test_size=.2, random_state=42, stratify=y)
```

```python
# Import standard scaler

scaler=StandardScaler()
```

```python
numeric_cols = [
    "alcohol_consumption_per_week",
    "physical_activity_minutes_per_week",
    "diet_score",
    "sleep_hours_per_day",
    "screen_time_hours_per_day",
    "bmi",
    "waist_to_hip_ratio",
    "systolic_bp",
    "diastolic_bp",
    "heart_rate",
    "cholesterol_total",
    "hdl_cholesterol",
    "ldl_cholesterol",
    "triglycerides",
    "glucose_fasting",
    "glucose_postprandial",
    "insulin_level",
    "hba1c"
]
```

```python
scaler.fit(X_train[numeric_cols])
```

```python
X_train[numeric_cols]=scaler.transform(X_train[numeric_cols])
X_test[numeric_cols]=scaler.transform(X_test[numeric_cols])
```

```python
X_train
```

```python
X_train=pd.get_dummies(X_train, drop_first=True)
X_test=pd.get_dummies(X_test, drop_first=True)
```
