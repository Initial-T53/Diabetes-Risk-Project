# Data Cleaning

The purpose of this notebook is to clean the data. The dataset came cleaned, however, Age and Gender need to be addressed.
I plan on binning age which improves data analysis by grouping together multiple like points, and removing the other category from gender.
For this analysis, the "Other" category from the gender variable. The dataset documentation does not specify whether "Gender" refers to gender identity or biological sex. Because this distinction is unclear, the variable lacks the specificity to interpret physiological differences relevant to diabetes risk. The "Other" category is particularly difficult to analyze because it is heterogeneous and cannot be mapped to a consistent biological construct. If the dataset had included a clearly defined "Sex" variable, it would have provided more meaningful information for examining biological factors. Given this ambiguity, excluding "Other" is a methodologically appropriate choice and will be documented accordingly.

After confirming that age has been dropped correctly, I will now focus on binning ages

```python
diabetes['age'].max()
```

```python
bins=[18,45,65,75,85, float('inf')]
labels=['18-44','45-64','65-74','75-84','85+']

diabetes['age']=pd.cut(diabetes['age'], bins=bins, labels=labels, right=False)
```

```python
diabetes['age'].value_counts()
```

```python
data_path = Path("..", "Dataset")

diabetes.to_csv(data_path / "diabetes_cleaned.csv", index=False)
```
