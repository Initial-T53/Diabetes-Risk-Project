```python
# import libraries

import pandas as pd
import numpy as np
import shap

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
# Load notebooks
X_train=pd.read_csv('../dataset/X_train.csv')
X_test=pd.read_csv('../dataset/X_test.csv')
y_train=pd.read_csv('../dataset/y_train.csv')
y_test=pd.read_csv('../dataset/y_test.csv')
```

Running the datasets to ensure they all loaded properly

# Logistic Regression

```python
# Training the logistic regression model using the training dataset
log_reg=LogisticRegression(
    max_iter=1000,
    solver='lbfgs')

log_reg.fit(X_train, y_train)
```

```python
y_pred=log_reg.predict(X_test)

y_proba=log_reg.predict_proba(X_test)[:,1]
```

```python
print(classification_report(y_test,y_pred))
```

Based on the classification report the overall accuracy of the model is 86% meaning 86% of all predictions are correct. Comparing the model for 0 (Patients not diagnosed with diabetes) and 1 (Patients diagnosed with diabetes) the model does a better job at correctly labeling a patient diabetic compared to label a patient non diabetic. When the model predicts a patient is diabetic 88% of the time the model is correct. While the model is 84% correct at predicting a patient does not have diabetes.

The model also does a good job at correctly identifing patients with diabetes as 89% of the patients it identities as diabetic are diabetic. While the model doesn't have as good of performance with an 81% correct rate for non diabetic rate.

The model has a higher f1-score for diagnosing diabetic patients at 89%. Meaning that it has a strong balance between catching cases and avoiding false alarms compared to the models ability to balance the precision and recall for not diabetic patients

```python
roc_auc=roc_auc_score(y_test, y_proba)
print("ROC-AUC", roc_auc)
```

Based on the high ROC-AUC score of 93.45% the model does a good job at distinguishing between individuals with and without diabetes across a wide range of classifications. Making it well suited for screening and risk straitication.

```python
confusion_matrix(y_test,y_pred)
```

Looking at the confusion matrix the model keeps the falses down to a minimum. The main category of concern is the false negatives. Which the model incorrectly says a patient doesn't have diabetes when they actually do have diabetes. The model has a false negative rate of 0.076 or 7.6%.

# Random Forest

```python
# Creating Randomforestclassifier
rf= RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    random_state=42,
    n_jobs=1)

rf.fit(X_train,y_train)
```

```python
y_pred_rf=rf.predict(X_test)
y_proba_rf=rf.predict_proba(X_test)[:,1]
```

```python
print(classification_report(y_test,y_pred_rf))
```

The random forest classification performed better than the logistic regression classifier. However, I would be lying to say I am not skeptical of the two 1.00 for precision and recall.

The model is 6% more accurate compared to the logistic regression model.  The model is great at identifying patients that don't have diabetes with 100% recall but is worst then the logistic regression model at identifying diabetic patients with a 87% recall.  The model is predicts non diabetic patients at the same level as logistic regression but has a 100% precision when predicting diabetic patients.

Random forest has 9% improvement compared to the logistic regression for non diabetic patients and a 4% improvement for diabetic patients

```python
roc_auc_rf=roc_auc_score(y_test, y_proba_rf)
print("RF ROC-AUC", roc_auc_rf)
```

Random forest classification also has a high ROC-AUC score with 94.51% The model does a really good job at distinguishing individuals that have and don't have diabetes

```python
confusion_matrix(y_test,y_pred_rf)
```

The confusion matrix data is really impressive for the Random Forest Classifier. The fact that the model only had 2 false negatives is really impressive. With the main category of concern being false negatives.

# XG Boost

```python
xgb=XGBClassifier(n_estimators=300,
                  learning_rate=0.05,
                  max_depth=6,
                  sub_sample=0.8,
                  colsample_bytree=0.8,
                  random_state=42,
                  eval_metric='logloss')

# Fitting the training data to XGBoost classifier

xgb.fit(X_train, y_train)
```

```python
y_pred_xgb=xgb.predict(X_test)
y_proba_xgb=xgb.predict_proba(X_test)[:,1]
```

```python
print(classification_report(y_test,y_pred_xgb))
```

```python
roc_auc_XGB=roc_auc_score(y_test, y_proba_xgb)
print("XGB ROC-AUC", roc_auc_XGB)
```

```python
confusion_matrix(y_test,y_pred_xgb)
```

Comparing the XGBoost Classifier to the Random forest classifier the results are almost the same. THe main difference comes in the ROC-AUC score where XGBoost performs slightly better compared to the Random Forest Classifier. However, the random forest classifer performs slightly better at identifying false positives with random forest classifier only have 2 false positives and XG Boost having 3. Because of the importance of correctly identifying patients, the best model is the Random forest classifier as it only incorrectly diagnosed negative patients twice instead of three times.

```python
# Changing the dtypes. Shaps dones't accept bools or objects

X_train_shap = X_train.astype(float)
X_sample_shap = X_sample.astype(float)
```

```python
explainer = shap.Explainer(rf, X_train_shap)
shap_values = explainer(X_sample_shap)

shap.summary_plot(shap_values[..., 1], X_sample_shap)
```

Looking at the shap summary plot. Hba1c, glucose fasting, glucode postprandial and family history of diabetes show the highest impact. There are also things like a patient having low physically activity impacting diabetes (Habitual)

# Patient Risk Categorizer

```python
# Creating risk labels
def risk_category(prob):
    if prob < 0.33:
        return "Low Risk"
    elif prob <0.66:
        return "Medium Risk"
    else:
        return "High Risk"
```

```python
# flatting y_test

y_test_f=y_test['diagnosed_diabetes'].values
```

```python
risk_labels=[risk_category(p) for p in y_proba_rf]

results= pd.DataFrame({'y_true':y_test_f,
                       'y_proba' : y_proba_rf,
                       'risk_category' : risk_labels})

results['risk_category'].value_counts()
```

After creating the algorithm, therer are 10227 high risk patients (Meaning they are high risk of having diabetes or having diabetes. There are 8945 patients that are at low risk of diabetes and 426 patients at medium risk of diabetes.

The most suprising part is how few medium risk patients there are. They are either high risk or low risk
